TOP 10 juegos mas Vendidos

¿Cuáles son los 10 videojuegos más vendidos de la historia, y cuál es su plataforma? Esta consulta busca identificar los videojuegos más populares de todos los tiempos, con el objetivo de analizar la evolución de la industria de los videojuegos a lo largo de los años

```python
top_game_seller = pd.read_sql_query('SELECT game, platform, total_shipped as sales \
                                     FROM game_sales \
                                     GROUP BY game \
                                     ORDER BY total_shipped DESC \
                                     LIMIT 10', conn)
df_top_game_seller = pd.DataFrame(top_game_seller)
df_top_game_seller
```

Nota: Agrupar los resultados por game y platform garantiza que si hay juegos con el mismo nombre para diferentes plataformas estos sean contabilizados por separados.

¿Cuáles son las 10 plataformas de videojuegos más populares en términos de ventas a lo largo de la historia? Esta consulta busca identificar las plataformas más exitosas en la industria de los videojuegos, con el objetivo de analizar las tendencias en ventas y determinar qué plataformas han sido las más influyentes en la historia de los videojuegos

```python
top_platform_seller = pd.read_sql_query('SELECT platform, SUM(total_shipped) AS total_copias \
                                        FROM game_sales \
                                        GROUP BY platform \
                                        ORDER BY total_copias DESC \
                                        LIMIT 10', conn)
df_top_platform_seller = pd.DataFrame(top_platform_seller)
df_top_platform_seller
```

Descubre los años dorados de la historia de los videojuegos. Obtén los 10 mejores años en ventas y adéntrate en una época donde los videojuegos alcanzaron su máximo esplendor. ¡Viaja en el tiempo y descubre qué años fueron los más exitosos para esta industria!

```python
top_year_seller= pd.read_sql_query('SELECT year, SUM(total_shipped) as copias_vendidas \
                                   FROM game_sales \
                                   GROUP BY year \
                                   ORDER BY copias_vendidas DESC \
                                   LIMIT 10', conn)
df_top_year_seller = pd.DataFrame(top_year_seller)
df_top_year_seller
```

Historico de Ventas por Año, fue creciendo o disminuyendo las ventas de los video juegos a lo largo de los años, las noticias señalan que ha disminuido, pero descubramos esto nosotros mismos y miremos que factor podria estar relacionado con esto?

```python
history_seller_year = pd.read_sql_query('SELECT year, COUNT(game) as lanzamientos, SUM(total_shipped) as copias_vendidas\
                                        FROM game_sales \
                                        GROUP BY year \
                                        ORDER BY year', conn)
df_history_seller_year = pd.DataFrame(history_seller_year)
df_history_seller_year
```

Después de realizar un análisis exhaustivo del historial de ventas de videojuegos por año, se pudo constatar que, efectivamente, las ventas de videojuegos han disminuido en los últimos años. Sin embargo, al examinar la cantidad de lanzamientos de videojuegos por año, se pudo observar que esta también ha disminuido significativamente en los mismos años en los que se vendieron menos juegos. Esto sugiere que la disminución en las ventas podría estar relacionada con la disminución en la cantidad de lanzamientos de videojuegos. No obstante, se debe continuar investigando porque es posible que se hayan lanzado menos juegos porque las plataformas se anticiparon a una temporada de ventas bajas

Ahoa hagamos mas pequeña nuestra busqueda tomando en cuenta las plataformas mas populares en los videos juegos

Como se comportaron las ventas para las plataformas mas populares en el periodo 2010 y 2020? hay que tomar en cuenta que hay plataformas que han prevalecido desde antes y durante todo este periodo y hay otras como Nientendo Switch que nacio casi al finalizar el peridodo

```python
seller_platform_per_year = pd.read_sql_query('SELECT platform, year, COUNT(game), SUM(total_shipped) as sales \
                                             FROM game_sales \
                                             WHERE platform IN (\'PS4\', \'XOne\', \'PC\', \'NS\') AND year BETWEEN 2010 AND 2020 \
                                             GROUP BY platform, year \
                                             ORDER BY platform, year', conn)
df_seller_platform_per_year = pd.DataFrame(seller_platform_per_year)
df_seller_platform_per_year
```

Llego en el momento de mirar los numeros de manera individual para cada protagonista

Como evolucionaron las ventas para Playstation para cada año de lanzamiento en un periodo comprendido entre 2010 y 2020

```python
ps_seller_year = pd.read_sql_query("SELECT year, COUNT(game) as releases, SUM(total_shipped) AS Playstation_sales \
                                  FROM game_sales \
                                  WHERE platform LIKE 'PS%' AND year BETWEEN 2010 AND 2020 \
                                  GROUP BY year \
                                  ORDER BY year", conn)
df_ps_seller_year = pd.DataFrame(ps_seller_year)
df_ps_seller_year
```

Como evolucionaron las ventas para Xbox  para cada año de lanzamiento en un periodo comprendido entre 2010 y 2020, habra sido un competidor fuerte para Playstation como siempre se rumora?

```python
xbox_seller_year = pd.read_sql_query("SELECT year, COUNT(game) as releases, SUM(total_shipped) AS Xbox_sales \
                                  FROM game_sales \
                                  WHERE platform LIKE 'X%' AND year BETWEEN 2013 AND 2020 \
                                  GROUP BY year \
                                  ORDER BY year", conn)
df_xbox_seller_year = pd.DataFrame(xbox_seller_year)
df_xbox_seller_year
```

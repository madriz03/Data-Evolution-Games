TOP 10 juegos mas Vendidos

¿Cuáles son los 10 videojuegos más vendidos de la historia, y cuál es su plataforma? Esta consulta busca identificar los videojuegos más populares de todos los tiempos, con el objetivo de analizar la evolución de la industria de los videojuegos a lo largo de los años

¿Cuáles son las 10 plataformas de videojuegos más populares en términos de ventas a lo largo de la historia? Esta consulta busca identificar las plataformas más exitosas en la industria de los videojuegos, con el objetivo de analizar las tendencias en ventas y determinar qué plataformas han sido las más influyentes en la historia de los videojuegos

```python
top_platform_seller = pd.read_sql_query('SELECT platform, SUM(total_shipped) AS sales \
                                        FROM game_sales \
                                        GROUP BY platform \
                                        ORDER BY sales DESC \
                                        LIMIT 10', conn)
df_top_platform_seller = pd.DataFrame(top_platform_seller)
df_top_platform_seller
```

```python
df_top_platform_seller.to_csv('./df_resultantes/top_platform_seller.csv', index=False)
```

```python
top_year_seller= pd.read_sql_query('SELECT year, SUM(total_shipped) as sales \
                                   FROM game_sales \
                                   GROUP BY year \
                                   ORDER BY sales DESC \
                                   LIMIT 10', conn)
df_top_year_seller = pd.DataFrame(top_year_seller)
df_top_year_seller
```

Historico de Ventas y lanzamientos de video juegos por Año, fue creciendo o disminuyendo las ventas de los video juegos a lo largo de los años, las noticias señalan que ha disminuido, pero descubramos esto nosotros mismos y miremos que factor podria estar relacionado con esto?

```python
history_seller_year = pd.read_sql_query('SELECT year, COUNT(game) as lanzamientos, SUM(total_shipped) as copias_vendidas\
                                        FROM game_sales \
                                        GROUP BY year \
                                        ORDER BY year', conn)
df_history_seller_year = pd.DataFrame(history_seller_year)
df_history_seller_year
```

```python
df_history_seller_year.to_csv('./df_resultantes/history_seller_year.csv', index=False)
```

Después de realizar un análisis exhaustivo del historial de ventas de videojuegos por año, se pudo constatar que, efectivamente, las ventas de videojuegos han disminuido en los últimos años. Sin embargo, al examinar la cantidad de lanzamientos de videojuegos por año, se pudo observar que esta también ha disminuido significativamente en los mismos años en los que se vendieron menos juegos. Esto sugiere que la disminución en las ventas podría estar relacionada con la disminución en la cantidad de lanzamientos de videojuegos. Llego el momento de revisar las calificaciones de los videos juegos, cuales son los video juegos mas querido, pertenecen a una misma plataforma? cual es el promedio de reviews de las plataformas mas populares?

TOP 10 de video juegos con mayor puntuacion de reviews, los mas queridos por los usuarios en la historia de los videos juegos

```python
top_game_reviews = pd.read_sql_query('SELECT gs.game, rv.critic_score AS max_score \
                                     FROM game_sales as gs \
                                     INNER JOIN reviews as rv\
                                     ON gs.game = rv.game \
                                     GROUP BY gs.game, rv.critic_score \
                                     ORDER BY max_score DESC \
                                     LIMIT 10', conn)
df_top_game_reviews = pd.DataFrame(top_game_reviews)
df_top_game_reviews
```

```python
df_top_game_reviews.to_csv('./df_resultantes_reviews/top_game_reviews.csv', index=False)
```

TOP 10 video juegos en la ultima decada. La ultima decada siempre es llamativo revisarla ya que con los avances de las tecnologias los juegos se hacen mas llamativos y entretenidos, revisemos el top 10 de los videos juegos mejores puntuados entre el 2010 y 2020

```python
top_game_reviews_lastdecade= pd.read_sql_query('SELECT gs.game, rv.critic_score AS max_core \
                                               FROM game_sales AS gs \
                                               INNER JOIN reviews AS rv \
                                               ON gs.game = rv.game \
                                               WHERE gs.year BETWEEN 2010 AND 2020 \
                                               GROUP BY gs.game, rv.critic_score \
                                               ORDER BY rv.critic_score DESC \
                                               LIMIT 10', conn)
df_top_game_reviews_lastdecade = pd.DataFrame(top_game_reviews_lastdecade)
df_top_game_reviews_lastdecade
```

```python
df_top_game_reviews_lastdecade.to_csv('./df_resultantes_reviews/top_game_reviews_lastdecade.csv', index=False)
```

TOP 10 Plataformas en la hisoria de lo videos juego, cuales fueron las plataformas con mejores reviews?

```python
top_platform_reviews = pd.read_sql_query('SELECT gs.platform, ROUND(AVG(rv.critic_score), 2) AS max_score \
                                         FROM game_sales AS gs \
                                         INNER JOIN reviews AS rv \
                                         ON gs.game = rv.game \
                                         GROUP BY gs.platform \
                                         HAVING COUNT(rv.critic_score)> 4 \
                                         ORDER BY max_score DESC \
                                         LIMIT 10', conn)
df_top_platform_reviews = pd.DataFrame(top_platform_reviews)
df_top_platform_reviews
```

```python
df_top_platform_reviews.to_csv('./df_resultantes_reviews/top_plaform_reviews.csv', index=False)
```

TOP 10 para plataformas con mejores reviews en la ultima decada. Bueno numeros para las plataformas mas populares, pero achiquemos un poco la busqueda a la ultima decada ya que la mision de nuestro analisis es ver el comportamiento de los video juegos es los tiempos mas recientes.

```python
top_platform_reviews_lastdecade = pd.read_sql_query('SELECT gs.platform, ROUND(AVG(rv.critic_score), 2) AS max_score \
                                                    FROM game_sales AS gs \
                                                    INNER JOIN reviews AS rv \
                                                    ON gs.game = rv.game \
                                                    WHERE gs.year BETWEEN 2010 AND 2020 \
                                                    GROUP BY gs.platform \
                                                    HAVING COUNT(critic_score) > 4 \
                                                    ORDER BY max_score DESC \
                                                    LIMIT 10', conn)
df_top_platform_reviews_lastdecade = pd.DataFrame(top_platform_reviews_lastdecade)
df_top_platform_reviews_lastdecade
```

```python
df_top_platform_reviews_lastdecade.to_csv('./df_resultantes_reviews/top_platform_reviews_lastdecade.csv', index=False)
```
